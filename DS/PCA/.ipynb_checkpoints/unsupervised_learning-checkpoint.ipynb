## 비지도 학습 이란?

- 대규모의 데이터들은 레이블이 붙어있는 경우가 많지 않습니다. 어떤 연속적인 데이터의 분포를 통해 유의미한 관계를 도출시킬 때, 혹은 데이터의 구조와 값을 연구해 결과를 도출시킬 수 있다. 비지도 학습은 주로 이런 상관 관계를 분석하는 데에 사용 됩니다.  

### 데이터로 부터 잠재 변수의 존재 파악하기

- 평일 오전 제주시에 사는 주민들이, 시내 버스를 이용할 것인지 추측해보자. 일정한 시간 대에 버스로 출/퇴근 하거나 등교하는 주민들 이라면, 시내 버스를 이용하는 빈도수에 대해서는 통계적으로 추론하기가 쉽다. 그러나, 늦잠을 자서 평소보다 늦은 시간대에 나왔다 던지, 전동 킥보드를 사용하고 출근하는 경우가 생긴다 던지 이런 가능성은 추론을 해야하는 변수로 데이터 셋에 반영이 되어 있지 않는다. 따라서 이를 다음과 같은 용어로 정의 할 수 있다. 
  

    - **관찰 변수** : 일정한 시간 대에 버스로 출/퇴근 하거나 등교하는 주민들의 수
    - **잠재 변수** : 늦잠을 자서 평소보다 늦은 시간대에 나오게 되는 수, 버스를 대체할 다른 교통 수단을 이용하게 되는 수  
      

- 이와 같이 데이터 셋의 내용을 근거로, 잠재변수와 변수의 값을 추론하는 과정이 매우 중요하다. 그 이유는 다음과 같다.  
  
    - 이러한 잠재 변수들은 데이터 셋에 이미 존재하는 변수들로 대체가 가능하다.
    - 데이터 셋에서 변수의 수를 줄임으로써, 데이터 셋이 관리하기 쉬워 지고 그로 인해 알고리즘이 더 빠르게 동작하며 정확도 또한 증가될 수 있다.
